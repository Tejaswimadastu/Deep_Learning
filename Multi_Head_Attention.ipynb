### Self Attention & Multi-Head Attention

```python
import tensorflow as tf
import numpy as np

from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import Input,Embedding,MultiHeadAttention,GlobalAveragePooling1D,Dense
from tensorflow.keras.models import Model
```

### Dataset

```python
sentences = [
"The animal didn't cross the street because it was tired",
"The trophy doesn't fit in the suitcase because it is too large",
"The book was placed on the table because it was heavy",
"The laptop was returned because it was damaged",
"The car stopped because it ran out of fuel"
]

labels = [0,1,2,3,4]
```

### Tokenization

```python
tokenizer=Tokenizer()
tokenizer.fit_on_texts(sentences)
sequences=tokenizer.texts_to_sequences(sentences)
x=pad_sequences(sequences,padding='post')
y=np.array(labels)
print(x.shape)
```

### Input Layer

```python
inputs=Input(shape=(x.shape[1],))
```

### Embedding Later

```python
embedding=Embedding(
    input_dim=len(tokenizer.word_index)+1,
    output_dim=32,

)(inputs)
```

### Multihead self Attention

```python
attention=MultiHeadAttention(
    num_heads=2,
    key_dim=32
)(
    query=embedding,
    key=embedding,
    value=embedding,
    return_attention_scores=True
)
```

### Pooling
Pooling converts:
(batch_size,sequence_length,embedding)

```python
pooled=GlobalAveragePooling1D()(attention[0])
```

### Dense Layers

```python
dense=Dense(64,activation='relu')(pooled)
outputs=Dense(5,activation='softmax')(dense)
```

### Build Model

```python
model=Model(inputs,outputs)
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy']


              )
model.summary()
```

```python
history=model.fit(
    x,y,epochs=50,verbose=1
)
```

### Prediction

```python
test_sentence = [
    "The animal didn't cross the street because it was tired"
]

seq = tokenizer.texts_to_sequences(test_sentence)
seq = pad_sequences(seq,maxlen=x.shape[1],padding='post')
prediction = model.predict(seq,verbose=0)
predicted_class=np.argmax(prediction)
print(predicted_class)
```

### Label Mapping

```python
label_mapping = {
    0:'animal',
    1:'trophy',
    2:'book',
    3:'laptop',
    4:'car'
}
print(label_mapping[predicted_class])
```
