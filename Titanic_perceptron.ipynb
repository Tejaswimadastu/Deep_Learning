```python
import numpy as np
x1 = 0.2
x2 = 0.24
x3 = 0.80
```

```python
target=1
lr = 0.1
```

```python
# Initial weights
w1,w2,w3 = 0.11,0.14,0.17
w4,w5,w6 = 0.21,0.24,0.27
w7,w8 = 0.31,0.34

# Biases
bh1 = 0.1
bh2 = 0.1
bo = 0.1
```

```python
# Sigmoid function
def sigmoid(x):
    return 1/(1+np.exp(-x))
```

# Forward propagation

```python
# Hidden layer
net_h1 = x1*w1 + x2*w2 + x3*w3 + bh1
net_h2 = x1*w4 + x2*w5 + x3*w6 + bh2

h1 = sigmoid(net_h1)
h2 = sigmoid(net_h2)

# Output layer
net_o = h1*w7 + h2*w8 + bo
output = sigmoid(net_o)

print("Hidden neuron h1:",h1)
print("Hidden neuron h2:",h2)
print("Predicted Output:",output)
```

#Error Calculation

```python
error = 0.5*(target-output)**2

print("Error:",error)
```

# Backpropagation

```python
delta_o = (target-output)*output*(1-output)

delta_h1 = h1*(1-h1)*(w7*delta_o)
delta_h2 = h2*(1-h2)*(w8*delta_o)
```

# Update weights

```python
w7 += lr*delta_o*h1
w8 += lr*delta_o*h2

w1 += lr*delta_h1*x1
w2 += lr*delta_h1*x2
w3 += lr*delta_h1*x3

w4 += lr*delta_h2*x1
w5 += lr*delta_h2*x2
w6 += lr*delta_h2*x3

# Update biases
bo += lr*delta_o
bh1 += lr*delta_h1
bh2 += lr*delta_h2

print("\nUpdated Weights")
print(w1,w2,w3)
print(w4,w5,w6)
print(w7,w8)

print("\nUpdated Biases")
print(bh1,bh2,bo)
```
