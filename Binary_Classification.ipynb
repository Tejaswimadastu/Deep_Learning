```python
import numpy as np
```

### Step 1: Prepare Training data

```python
#Training data

X = np.array([1,2,3,4]) #Study hours
y = np.array([0,0,1,1]) #Pass(1)/Fail(0)

X,y
```

### Step 2: Initialize Parameters

```python
w = 0.0
b = 0.0
learning_rate = 1
epochs = 5

print("Initial Weight : ",w)
print("Inintial Bias : ",b)
```

### Define Activation Function (Step Function)

```python
def step_function(z):
    return 1 if z>=0 else 0
```

### Training Loop

```python
for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    for i in range(len(X)):
        # 1. Compute Weighted sum
        z = w * X[i] + b

        # 2. Apply Activation Function
        y_pred = step_function(z)

        # 3. Compute Error
        error = y[i] - y_pred

        # 4. Update Weight
        w = w + learning_rate * error * X[i]

        # 5. Update Bias
        b = b + learning_rate * error

        print(f"Input ; {X[i]}, Predicted : {y_pred}, Actual : {y[i]}, w : {w}, b : {b}")
```

### Final Model parameters

```python
print("Final Weight ",w)
print("Final bias ",b)
```

# Testing Model

```python
test_value = 2.9
z = w * test_value + b
prediction = step_function(z)

print(f"For study hours {test_value}, Prediction {prediction}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate smooth x range across your data
x_vals = np.linspace(min(X)-1, max(X)+1, 200)

# Linear combination
z_vals = w * x_vals + b

# Step predictions
y_vals = [step_function(z) for z in z_vals]

# Plot
plt.figure()
plt.scatter(X, y)          # training points
plt.plot(x_vals, y_vals)   # perceptron step curve
plt.xlabel("Study Hours")
plt.ylabel("Prediction")
plt.title("Perceptron Decision Boundary")
plt.show()
```
