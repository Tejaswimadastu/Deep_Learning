# Tokenization
breaking text into smaller units called tokens

🔸 Types of Tokens:

Word Tokenization

Sentence → words
👉 "I love AI" → ["I", "love", "AI"]

Sentence Tokenization

Paragraph → sentences
👉 "Hello. How are you?" → ["Hello.", "How are you?"]

Character Tokenization

Word → characters
👉 "AI" → ["A", "I"]

```python
from nltk.tokenize import word_tokenize
text="I love Learning Natural Language Processing"
#word level
tokens=word_tokenize(text)
print(tokens)

#char level
tokens=list(text)
print(tokens)

```

# Stopword Removal

Examples:
is, the, a, an, and, of

Example:

Tasks that benefit from stopword removal:

Information retrieval and search engines

Topic modelling and clustering

Keyword extraction

Tasks that require preserving stopwords:

Machine translation (maintains grammatical structure)

Text summarization (preserves sentence coherence)

Question-answering systems (syntactic relationships matter)

Grammar checking and parsing

```python
from nltk.corpus import stopwords

stop_words = set(stopwords.words('english'))

print(stop_words)

print("After Stopword Removal")
filtered_tokens = [token for token in tokens if token not in stop_words]
print(filtered_tokens)
```

# Stemming

Stemming reduces words to their root form.

Stemming is an important text-processing technique that reduces words to their base or root form by removing prefixes and suffixes. This process standardizes words which helps to improve the efficiency and effectiveness of various natural language processing (NLP) tasks.

Examples of stemming for the word "like":

"likes" → "like"

"liked" → "like"

"likely" → "like"

"liking" → "like"

```python
from nltk.stem import PorterStemmer
stemmer=PorterStemmer()
words=["Running","Jumps","liked","running","playing","Happily"]
stemmed_words=[stemmer.stem(word) for word in words]
print("Original ",words)
print("Stemmed ",stemmed_words)
```

# Snowball Stemmer:
is a advanced version of Porter Stemmer, Supports multiple languaues

```python
from nltk.stem import SnowballStemmer
snowball_stemmer=SnowballStemmer(language='english')
words_to_stem = ["running", "jumps", "liked", "running", "playing", "happily"]
stemmed_words = [snowball_stemmer.stem(word) for word in words_to_stem]
print("Original ",words_to_stem)
print("Stemmed ",stemmed_words)
```

# Lancaster Stemmer:
is known for being more aggressive and faster than other stemmers.

```python
from nltk.stem import LancasterStemmer

lancaster_stemmer = LancasterStemmer()

words_to_stem = ["running", "jumped", "Happily", "Quickly", "Foxes"]

stemmed_words = [lancaster_stemmer.stem(word) for word in words_to_stem]

print("Original Words:", words_to_stem)
print("Stemmed Words:", stemmed_words)
```

# Lemmatization

Lemmatization produces meaningful root words.

**Example:**
better → good
running → run


Lemmatization is an important text pre-processing technique in Natural Language Processing (NLP) that reduces words to their base form known as a "lemma". For example, the lemma of "running" is "run" and "better" becomes "good".


Unlike stemming which simply removes prefixes or suffixes, it considers the word’s meaning and part of speech (POS) and ensures that the base form is a valid word. This makes lemmatization more accurate as it avoids generating non-dictionary words.

**Used for:**

Improves accuracy
Reduced Data Redundancy
Better NLP Model Performance

```python
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

print(lemmatizer.lemmatize("better", pos='a'))
print(lemmatizer.lemmatize("running", pos='v'))
```

Text Representation Techniques

Machines cannot understand text.
They understand numbers.
So we convert text → numbers.

Bag of Words (BoW)

BoW represents text as word frequency vectors.

In Natural Language Processing (NLP) text data needs to be converted into numbers so that machine learning algorithms can understand it. One common method to do this is Bag of Words (BoW) model. It turns text like sentence, paragraph or document into a collection of words and counts how often each word appears but ignoring the order of the words. It does not consider the order of the words or their grammar but focuses on counting how often each word appears in the text.

This makes it useful for tasks like text classification, sentiment analysis and clustering.

# Step1 : Preprocessing the text

clean text

covert to lower

remove special characters

```python
import re

dataset = []

text = [
    "I love coding!",
    "I love Python's",
    "Python coding is fun"
]

for sentence in text:
    sentence = sentence.lower()              # Lower case
    sentence = re.sub(r'\W', ' ', sentence)  # Remove symbols
    sentence = re.sub(r'\s+', ' ', sentence) # Remove extra spaces
    dataset.append(sentence)

print(dataset)
```

# Step 2: Counting Word Frequencies

Count how many times each word appears

```python
from collections import defaultdict

wordCount = defaultdict(int)

for sentence in dataset:
    for word in sentence.split():
        wordCount[word] += 1
print(wordCount)
```

# Selecting most frequent words



```python
import heapq

freq_words = heapq.nlargest(3, wordCount, key=wordCount.get)
print(freq_words)
```

# ** Step 4: Build Bag of Words Matrix**

Now we create a matrix:

👉 Rows = sentences
👉 Columns = frequent words
👉 Value = 1 (present) or 0 (absent)

```python
x=[]
for sentence in dataset:
  vector=[]
  words=sentence.split()
  for word in freq_words:
    if word in words:
      vector.append(1)
    else:
      vector.append(0)
  x.append(vector)
print(x)
```

# N-grams

N-grams capture word sequences.

N-gram is a technique that splits text into sequences of N consecutive words to capture context and word order for machine learning models.

N-gram = group of N consecutive words

1 word → Unigram
2 words → Bigram
3 words → Trigram

✅ Simple Example

Sentence:
"I love coding"

Unigram (N = 1)

["I", "love", "coding"]

Bigrams (N = 2)

["I love", "love coding"]

Trigrams (N = 3)

["I love coding"]

```python
# preprocessing text

import re
text={
    "I Love Coding",
    "I Love Python",
    "Python coding is fun"
}
dataset=[]
for sentence in text:
  sentence = re.sub(r'\W', ' ', sentence)
  sentence = re.sub(r'\s+', ' ', sentence)
  dataset.append(sentence)

print(dataset)
```

```python
# Step 2: Create N-grams
# Function to generate N-grams
def generate_ngrams(sentence, n):
    words = sentence.split()
    ngrams = []
    for i in range(len(words) - n + 1):
        ngrams.append(' '.join(words[i:i+n]))
    return ngrams
```

```python
# Example: Bigrams (N=2)
for sentence in dataset:
    bigrams = generate_ngrams(sentence, 2)
    print(bigrams)

# Example: Trigrams (N=3)
for sentence in dataset:
    trigrams = generate_ngrams(sentence, 3)
    print(trigrams)
```

```python
# Step 3: Count N-gram Frequencies
from collections import defaultdict

ngramCount = defaultdict(int)

for sentence in dataset:
    ngrams = generate_ngrams(sentence, 2)
    for gram in ngrams:
        ngramCount[gram] += 1

print(ngramCount)
```

```python
# Step 4: Convert into feature vector (Like BoW)
# Select top N-grams:
import heapq

freq_ngrams = heapq.nlargest(2, ngramCount, key=ngramCount.get)
print(freq_ngrams)
```

```python

X = []

for sentence in dataset:
    vector = []
    sentence_ngrams = generate_ngrams(sentence, 2)

    for gram in freq_ngrams:
        if gram in sentence_ngrams:
            vector.append(1)
        else:
            vector.append(0)

    X.append(vector)

print(X)
```

# TF-IDF: (Term Frequency–Inverse Document Frequency)

TF-IDF gives importance to rare words. It is used to measure how important a word is in a document.

🤔 Why Do We Need TF-IDF?

Problem with BoW:

Words like "the", "is", "I" appear many times

But they are not important

👉 BoW treats all words equally ❌

```python
text=[
    "I Love Coding",
    "I Love Python",
    "Python is Powerful"
]
from sklearn.feature_extraction.text import TfidfVectorizer

vectorizer = TfidfVectorizer()
x=vectorizer.fit_transform(text)
print(vectorizer.get_feature_names_out())
print(x.toarray())
```

# Word Embeddings =
representing words as meaningful numbers (vectors)

Unlike BoW/TF-IDF:

They don’t just count words ❌

They capture meaning and relationships ✔

🤔 Why Do We Need Word Embeddings? ❌ Problem with TF-IDF / BoW

Example:

"good" and "excellent"

👉 BoW / TF-IDF:

Treat them as completely different ❌

✅ Word Embeddings Fix This

👉 They understand:

"good" ≈ "excellent" ✔

"king" - "man" + "woman" ≈ "queen" 👑

| Sentence | coding | love | python | powerful |
|----------|--------|------|--------|----------|
| D1       | 0.65   | 0.38 | 0      | 0        |
| D2       | 0      | 0.38 | 0.65   | 0        |
| D3       | 0      | 0    | 0.38   | 0.65     |

Two methods:

CBOW (predict word from context)

Skip-gram (predict context from word)
1. CBOW (Continuous Bag of Words)

🎯 Idea:
👉 Predict the center word using surrounding words

❌ Example

Sentence:
"I love coding"

👉 Input (context words):
"I" and "coding"

👉 Output (target word):
"love"

2. Skip-gram

🎯 Idea:
👉 Predict surrounding words using center word

❌ Example

Sentence:
"I love coding"

👉 Input (center word):
"love"

👉 Output (context words):
"I", "coding"

```python
!pip install gensim
```

```python
# CBOW
from gensim.models import Word2Vec

sentences = [
    ["i", "love", "coding"],
    ["i", "love", "python"],
    ["python", "is", "powerful"]
]

# Train CBOW model (sg=0)
cbow_model = Word2Vec(
    sentences,
    min_count=1,
    vector_size=5,   # size of word vectors
    window=2,        # context window
    sg=0             # 0 = CBOW
)
# Get Word Vector
print("Vector for coding:")
print(cbow_model.wv["coding"])

# Find similar words
print("Similar to coding:")
print(cbow_model.wv.most_similar("coding"))
```

```python
# skip-gram
from gensim.models import Word2Vec

sentences = [
    ["i", "love", "coding"],
    ["i", "love", "python"],
    ["python", "is", "powerful"]
]

# Train CBOW model (sg=0)
cbow_model = Word2Vec(
    sentences,
    min_count=1,
    vector_size=5,   # size of word vectors
    window=2,        # context window
    sg=1
)
# Get Word Vector
print("Vector for coding:")
print(cbow_model.wv["coding"])

# Find similar words
print("Similar to coding:")
print(cbow_model.wv.most_similar("coding"))
```

🎯 What is NER?

👉 Named Entity Recognition (NER) is the process of identifying and classifying important words in a sentence into categories like:

Person 👤

Location 🌍

Organization 🏢

Date 📅

Money 💰

Example

Sentence:
"Elon Musk founded Tesla in California in 2003"

NER Output:

Elon Musk → PERSON

Tesla → ORGANIZATION

California → LOCATION

2003 → DATE

```python
sentence = "Elon visited India"

words = nltk.word_tokenize(sentence)
pos_tags = nltk.pos_tag(words)

ner_tree = nltk.ne_chunk(pos_tags)

print(ner_tree)
```
