```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([0,1,1,0])

w = np.zeros(2)
b = 0
lr = 0.1
```

```python
def step(z):
    return 1 if z>=0 else 0

for _ in range(10):
    for i in range(len(X)):
        z = np.dot(w,X[i]) + b
        y_hat = step(z)
        w += lr*(y[i]-y_hat)*X[i]
        b += lr*(y[i]-y_hat)
```

```python
for x in X:
    print(x, step(np.dot(w,x)+b))
```

```python
import numpy as np

X = np.array([[0,0],[0,1],[1,0],[1,1]])
y = np.array([[0],[1],[1],[0]])

W1 = np.random.randn(2,2)
b1 = np.zeros((1,2))
W2 = np.random.randn(2,1)
b2 = np.zeros((1,1))

lr = 0.1
```

```python
def sigmoid(z):
    return 1/(1+np.exp(-z))

def dsigmoid(a):
    return a*(1-a)

for _ in range(10000):
    z1 = X@W1 + b1
    a1 = sigmoid(z1)
    z2 = a1@W2 + b2
    a2 = sigmoid(z2)

    dz2 = (a2 - y)*dsigmoid(a2)
    dW2 = a1.T@dz2
    db2 = np.sum(dz2,axis=0,keepdims=True)

    dz1 = dz2@W2.T * dsigmoid(a1)
    dW1 = X.T@dz1
    db1 = np.sum(dz1,axis=0,keepdims=True)

    W2 -= lr*dW2
    b2 -= lr*db2
    W1 -= lr*dW1
    b1 -= lr*db1
```

```python
pred = (a2>0.5).astype(int)
print(pred)
```
