```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,1])
```

```python
w = np.array([1,1])
b = -0.5
```

```python
def perceptron_predict(x):
    z = np.dot(w, x) + b
    if z >= 0:
        return 1
    else:
        return 0
```

```python
print("Predictions:")
for i in range(len(X)):
    print(X[i], "->", perceptron_predict(X[i]))
```

```python
plt.figure(figsize=(6,5))

for i in range(len(X)):
    if y[i]==0:
        plt.scatter(X[i,0], X[i,1], color='blue', label="Not Spam" if i==0 else "")
    else:
        plt.scatter(X[i,0], X[i,1], color='red', marker='x', s=100, label="Spam" if i==1 else "")
```

```python
x1 = np.linspace(-0.5,1.5,100)
x2 = -(w[0]*x1 + b)/w[1]
plt.plot(x1, x2, label="Decision Boundary")

plt.xlim(-0.5,1.5)
plt.ylim(-0.5,1.5)
plt.xlabel("Contains Keywords")
plt.ylabel("Unknown Sender")
plt.title("Startup Spam Filter (Perceptron)")
plt.grid(True)
plt.legend()

plt.show()
```
