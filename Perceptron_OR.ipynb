```python
import numpy as np
import matplotlib.pyplot as plt
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])
y = np.array([0,1,1,1])
```

```python
w = np.zeros(2)
b = 0
lr = 1
epochs = 10
```

```python
for _ in range(epochs):
    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        
        if z >= 0:
            y_pred = 1
        else:
            y_pred = 0
        
        error = y[i] - y_pred
        
        w = w + lr * error * X[i]
        b = b + lr * error

print("Learned weights:", w)
print("Learned bias:", b)
```

```python
def predict(x):
    z = np.dot(w, x) + b
    return 1 if z >= 0 else 0

# ======================
# 5. Test predictions
# ======================
print("\nPredictions:")
for i in range(len(X)):
    print(X[i], "->", predict(X[i]))
```

```python
correct = 0
for i in range(len(X)):
    if predict(X[i]) == y[i]:
        correct += 1

print("\nAccuracy:", correct/len(X))
```

```python
plt.figure(figsize=(6,5))

for i in range(len(X)):
    if y[i] == 0:
        plt.scatter(X[i,0], X[i,1], color='blue', label="Class 0" if i==0 else "")
    else:
        plt.scatter(X[i,0], X[i,1], color='red', marker='x', s=100, label="Class 1" if i==1 else "")
```
