```python
import numpy as np
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,0,0,1])
X,y
```

```python
w = np.zeros(2)
b = 0
learning_rate = 0.1
epochs = 10

print("Initial Weights:", w)
print("Initial Bias:", b)
```

```python
def step(z):
    return 1 if z >= 0 else 0
```

```python
for x in X:
    z = np.dot(w, x) + b
    y_pred = step(z)
    print(x, "->", y_pred)
```

```python
print("\nFinal Weight : ",w)
print("Final Bias : ",b)
```

# Train Perceptron

```python
for epoch in range(epochs):
    print(f"\n Epoch {epoch+1}")

    for i in range(len(X)):
        z = np.dot(w, X[i]) + b
        y_pred = step(z)
        error = y[i] - y_pred

        w = w + learning_rate * error * X[i]
        b = b + learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_pred}, Error: {error}")
```

# Test Model

```python
print("\nTesting AND Gate")

for i in range(len(X)):
    z = np.dot(w, X[i]) + b
    prediction = step(z)
    print("Input:", X[i], "Prediction:", prediction)
```
